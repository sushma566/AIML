```python
import matplotlib.pyplot as plt
```

```python
y=[1,2,3,4]
x=[10,20,30,40]
plt.plot(x,y)
```

```python
plt.plot(y)
```

```python
plt.plot(y,color='green',linestyle='--',marker='*',markersize=8,markerfacecolor='red',linewidth=2)
```

```python
x_1=[45,20,67,90]
plt.plot(x,y,marker='o')
plt.plot(x_1,y,marker="x")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")
plt.legend(["x","x_1"])
```

```python
x=[1,2,3,4]
y=[10,20,30,40]
x_1=[5,6,7,8]
plt.plot(x,y,marker="o")
plt.plot(x_1,y,marker="x",color='green')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")
plt.legend(["x","x_1"])

```

```python
year=[20,21,22,23,24]
money=[1,20,10,30,50]
plt.bar(year,money)
plt.xlabel("year")
plt.ylabel("money")
plt.title("profit")
plt.show()
```

```python
y=[0,0,.2,0,0]
money=[1,20,10,30,50]
plt.pie(money,explode=y)


```
